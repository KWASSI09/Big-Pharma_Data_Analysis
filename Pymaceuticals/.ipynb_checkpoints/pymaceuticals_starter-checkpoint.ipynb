## Observations and Insights 

The total number of mice used for drug treatment without duplicates is 248, with 125 males and 123 females thus represented in the pie chart an almost equal distribution with percentages of 50.4% and 49.6%, respectively. Regarding the timepoints observing the bar chart, Capomulin and Ramicane were used more in the drug regimen, with a total number of time points of 230 for the first and 228 for the last.


For the m957 mouse under treatment with Capomulin, it can be observed that the tumor volume grows relatively between 0 and 10-time points and then decreases dramatically. However, among the 10 drug regimens, treatment with Ramicane shows positive values in tumor reduction, having the lowest mean, median, var, std and sem.

The correlation and the regression line show a positive but non-linear relationship between tumor weight and volume. That is, there is a considerable dependence of the average volume of the tumor on the weight but it is not linear.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
mouse_metadata.head()
```

```python
study_results.head()
```

```python
# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_metadata,study_results, how = "outer", on = "Mouse ID")

# Display the data table for preview
combined_data_df.head()
```

```python
# Checking the number of mice.
mice_number = combined_data_df["Mouse ID"].nunique()
mice_number
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice_ID = combined_data_df.loc[combined_data_df.duplicated(subset=["Mouse ID","Timepoint"]),"Mouse ID"].unique()
duplicate_mice_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_df = combined_data_df.loc[combined_data_df["Mouse ID"]== "g989", :]
duplicate_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
noduplicate_df =  combined_data_df[ combined_data_df['Mouse ID'].isin(duplicate_mice_ID)==False]
noduplicate_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
real_mice_number = noduplicate_df["Mouse ID"].nunique()
real_mice_number
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean = noduplicate_df["Tumor Volume (mm3)"].groupby(noduplicate_df["Drug Regimen"]).mean()
median = noduplicate_df["Tumor Volume (mm3)"].groupby(noduplicate_df["Drug Regimen"]).median()
var= noduplicate_df["Tumor Volume (mm3)"].groupby(noduplicate_df["Drug Regimen"]).var()
std = noduplicate_df["Tumor Volume (mm3)"].groupby(noduplicate_df["Drug Regimen"]).std()
SEM = noduplicate_df["Tumor Volume (mm3)"].groupby(noduplicate_df["Drug Regimen"]).sem()
# Assemble the resulting series into a single summary dataframe.
summary_df = pd.DataFrame({ "Mean Tumor Vol":mean,
                          "Meadian Tumor Vol":median,
                          "Variance Tumor Vol":var,
                          "Std.Dev Tumor Vol":std,
                          "Std.Err Tumor Vol":SEM}).style.set_table_styles([{'selector': 'td', 'props': [('text-align', 'left')]},
                           {'selector': '.col_heading', 'props': [('text-align', 'center')]},
                           {'selector': '.row_heading', 'props': [('text-align', 'left')]}])
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
agg_sum = noduplicate_df.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean","median","var","std","sem"]).style.set_table_styles([{'selector': 'td', 'props': [('text-align', 'left')]},
                           {'selector': '.col_heading', 'props': [('text-align', 'right')]},
                           {'selector': '.row_heading', 'props': [('text-align', 'left')]}])
agg_sum
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
#Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
greatest_timepoint=noduplicate_df.groupby(['Mouse ID'])['Timepoint'].max()
greatest_timepoint
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tv_df = pd.merge(greatest_timepoint,noduplicate_df, on = (["Mouse ID","Timepoint"]))
tv_df
```

```python
# Put treatments into a list for loop (and later for plot labels)
#Capomulin, Ramicane, Infubinol, and Ceftamin
Drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

#Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Locate the rows which contain mice on each drug and get the tumor volumes for treatment in treatments: 
for drug in Drugs:
    tv_regimen = tv_df.loc[tv_df["Drug Regimen"] == drug]
    tumor_vol= tv_regimen['Tumor Volume (mm3)']
    
#append tumor data to new list 
    tumor_vol_data.append(tumor_vol)
    
    
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers - Capomulin
quartile_0 = round(tumor_vol_data[0].quantile([.25,.5,.75]),3)
lowerq_0 = round(quartile_0[0.25],3)
upperq_0 = round(quartile_0[0.75],3)
iqr_0 = round(upperq_0-lowerq_0,3)
lower_bound_0 = round(lowerq_0 - (1.5*iqr_0),3)
upper_bound_0 = round(upperq_0 + (1.5*iqr_0),3)

print(f"The lower quartile of {Drugs[0]} tumor volume: {lowerq_0}")
print(f"The upper quartile of {Drugs[0]} tumor volume: {upperq_0}")
print(f"The interquartile range of {Drugs[0]} tumor volume: {iqr_0}")
print(f"The median of {Drugs[0]} tumor volume:{quartile_0[0.5]}")
print(f"Values below {lower_bound_0}and above {upper_bound_0} could represent the outliers.")

```

## Correlation and Regression

```python
# Calculate the correlation and regression model between mouse weight and average tumor volume for the Capomulin treatment
correlation = st.pearsonr(avg_capo_vol["Weight (g)"],avg_capo_vol["Tumor Volume (mm3)"])
```

```python
# Add the linear regression equation and line to plot
# Calculate the correlation coefficient and linear regression model 
#for mouse weight and average tumor volume for the Capomulin regimen
(slope,intercept,rvalue,pvalue,stderr)=linregress(avg_capo_vol["Weight (g)"],avg_capo_vol["Tumor Volume (mm3)"])
regress_values = avg_capo_vol["Weight (g)"]* slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(avg_capo_vol["Weight (g)"],avg_capo_vol["Tumor Volume (mm3)"], color="blue")
plt.plot(avg_capo_vol["Weight (g)"], regress_values,"r-")
plt.annotate(line_eq,(20,35),fontsize=14)
plt.xlabel("Weight (g)",fontsize =12)
plt.ylabel("Avg Tumor Volume (mm3)",fontsize =12)
plt.title("average tumor volume vs. mouse weight for the Capomulin regimen",fontsize =14)
print(f"The correlation between mouse weight and the average tumor volume for the Capomulin regimen is {round(correlation[0],2)}")
print(f" the linear correlation coefficient of bravais pearson r-squared is :{round(rvalue**2,3)}")
plt.show()
```
